# STRING ATTRIBUTES

- Capitalize the first letter { s.capitalize}

- Lowers all the the letters {a.lower()}
- Give the all starting letter in uppercase {a.title()}
- give the space or include some character in the of specific value gievn in the parenthesis{a.center(21,#)}
- here count is use dto show the number of occurance of the specific character{a.count("c")}

```python
a = "wElCoMe To Cttc"
print(a.capitalize())
print(a.upper())
print(a.lower())
print(a.title())
print(a.swapcase())
print(a.casefold())
print(a.center(21,"#"))
print(a.count("C"))
print(a.find('C'))
```

# Strip 
_ _ _ 
* strip is used to remove or the space or the specific character given inside the parenthesis { var.strip(#)}
* rstrip is used to to remove the space and the specific character given inside the parenthesis from the RIGHT SIDE { var.rstrip(#)}
* lstrip is used to to remove the space and the specific character given inside the parenthesis from the LEFT SIDE { var.lstrip(#)}


```python

print('###wElCoMe To Cttc###'.strip('#'))
print('###wElCoMe To Cttc###'.rstrip('#'))
print('###wElCoMe To Cttc###'.lstrip('#'))
print('###wElCoMe To Cttc###'.index('e'))
print('###wElCoMe To Cttc###'.index('C'))
```

# Check for true or false
_ _ _ 
* var.isalpha() - is used for to check the element is only alphabets or not
* var.isalnum() - is for to check the element has both numeric and alphabets or not
* var.isnumeric() - is used to check the element contains only numeric value or not
* var.islower()- is used to check the all the characters of the element is in lowercase or not
* var.isupper()- is used to check the all the characters of the element is in uppercase or not
* 

```python
print('abcd'.isalpha())
print('abc25d'.isalpha())
print('abc22d'.isalnum())
print('4568'.isnumeric())
print('abcd'.islower())
print('abcd'.isupper())
```

```python
l=[45, 63.5, True,"False" ]
l.append(45)
print(l)

l.insert(2,22)

l.extend([12.5,'abc'])
print(l)

m=[56,29.5]
m.extend(l)
print(m)

l.count(45)
l.pop() #delete from the last
print(l)

m.remove(45)  #remove the first occurance element in the list
print(m)

m.clear()

print(l)

# l.remove('abc')

# l.sort()

l.reverse()
print(l)

# l.sort(reverse=True)
```

# TUPLE ATTRIBUTE
_ _ _
* type(t)  is used to show the data type
* t[4] is used to show the indexing 
* t[0:6:2] is used to print the element from index 0 to 6 by skipping two steps t[start: stop:steps]
* t.index(value) is used to print the index of the value inside the parenthesis in the tuple

_  _ _ end of tuple _ _ _ 

```python
#TUPLE attribute

t=(63.5, 45, 45, 22, 12.5, 12.5, 2, 2, True)
print(type(t))
print(t[4])

print(t[0:6:2])

print(t.count(45))

print(t.index(True))
```
