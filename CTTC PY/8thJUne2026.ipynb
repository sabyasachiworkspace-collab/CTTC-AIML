* write a python program to print the grade by taking mark of a student by using if-elif-else 
* <pre>
if marks> 90 = O
80> = E

```python
marks = int(input('Enter your marks: '))
if(marks>=90):
    print('O')
elif(marks>=80):
    print('E')
elif(marks>=70):
    print('A')
elif(marks>=60):
    print('B')
elif(marks>=50):
    print('C')
elif(marks>=40):
    print('D')
else:
    print('Fail')
  
```

Q write a py prgm using for loop to find the prime number between entered numbers

```python
n = int(input("Enter the initial number: "))
fact = 0
for i in range(1,n+1):
    if(n%i==0):
        fact+=1

if(fact==2):
    print(f'{n} is Prime')
else:
    print(f'{n}is Not Prime')
```

```python
#finding prime for singlle number using while
n = int(input('Enter a number: '))
fact = 0
i = 1
while(i<=n):
    if(n%i==0):
        fact+=1
    i+=1
    

if(fact==2):
    print("The number",n," is prime")
else:
    print("The number is not prime")
```
