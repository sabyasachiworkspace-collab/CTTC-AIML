```python
import pandas as pd
import numpy as np

df = pd.DataFrame({'Days':[1,2,3,4,5],  
                   'Places':['Puri', 'Mathura', 'Asansol','Manali','Kolkata'],
                   'Visitors':[5895,5600,4859, 5000, 5800] })
df
```

```python
# to check missing value
df.isnull()
# isnull().sum() is used to count the number of false or true
df.isnull().sum()
```

```python
df = pd.DataFrame({'Days':[1,2,3,4,5],  
                   'Places':['Puri', 'Mathura', np.nan,'Manali','Kolkata'],
                   'Visitors':[5895,5600,4859, 5000, 5800] })
df
df.isnull()
```

```python
# import warnings 
# warnings.filterwarnings('ignore')
```

```python
# fillna = to fill null value
df.Places.fillna('Puri')
df.Places.fillna('Puri',inplace=True)
df
```
