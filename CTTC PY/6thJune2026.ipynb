<pre><b>Types of operator:
    . Arithmetic operator [ +, -, /, *, %, //, **]

```python
#Arithmetic operators
a = 13
b = 3
print('a=', a)
print('b=', b)
print(a+b)
print('a+b= ', a+b)
print('a-b= ', a-b)
print('a*b= ', a*b)
print('a/b= ', a/b)
print('a%b= ', a%b)
print('a**b= ', a**b)
print('a//b= ', a//b)
```

```python
# Relational operators
print(5<10)
print(4!=5)
```

```python
# lOgical operators
print(5>10 or 5<10) # False or True = True
print(5>10 and 5<10) # False and True = False
print(not 50>6) # not + False = True similarly not + True = False
print(not 5>10 and 5<10) # not + False and True = True and True = True
```

```python
# Identity operators gives the value true or false if the both variable memory location is same or not
# if we take any number between -5 to 256 then the memory location of that number will be same but if we take any number outside that range then the memory location will be different
a = 10
b = 10  
print(a is b) # True
print(a is not b) # False
print(id(a)) # 140703353906512
print(id(b)) # 140703353906512
c= [1,2,3]
d= [1,2,3]  
print(id(c)) # 140703353906512
print(id(d)) # 140703353906512
print(c is d) # False
print(c == d) # True
e=500
f=500
print(e is f) # False because the memory location of e and f is different
print(e == f) # True    

x= 'Nirmal'
y = 'Nirmal'
print(x is y) # True
print(x == y) # True
w = 'Sabya'
z = 'Sabya'
print(w is z) # True
print(w == z) # True
```

```python
z = '_'.join(['Nirmal', 'kumar'])
x = 'Nirmal_kumar'
print(z)
print(x)
x is z # False because the memory location of x and z is different

x is not z # True because the memory location of x and z is different
```
