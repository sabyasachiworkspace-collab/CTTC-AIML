# Functions:
A reusable piece of code. Block of statements that does some specific task and returns asomething.

syntax: def function_name(args):
            # block of codes
        
        function_name()

```python
def greet():        #function declaration
    print('Hello')  #function definition
greet()             #function call
```

```python
# with one argument
def greet(name):
    print('Hello',name)
    
greet('Sabya')
```

```python
# with one argument and return
def greet(name):
    return 'Hello '+name
    
c = greet('Sabya')
print(c)
```

```python
# two argements
def add(num1, num2):
    return num1+num2

c = add(5,6)
print(c)
d = add(num2= 55, num1=20)
print(d)
```

```python
# default arguments
def add(m=5,n=10):
    return m+n
add(6,1)
```

```python
# multiple args;
def add(*arg):
    return sum(arg)
add(10,20,30)
```
